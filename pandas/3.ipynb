```python
data4={
    'emp_id': [101,102,103,104],
    'salary': [234,345,200,400],
    'pf':[23,34,56,78]
    
}
```

```python
df4=pd.DataFrame(data4)
df4
```

```python
data5={
    'emp_id':[101,102,104,105],
    'mob_no':[3434,5656,767,878],
    'house_nu':[345,345,345,56]
}
```

```python
df5=pd.DataFrame(data5)
df5
```

```python
# Merge operation and how is it different from aconcat operation

pd.merge(df4,df5)
# It strictly finds a common column and tries to merge rows based on that. It ignores any exclusive row data to the common column just like how it ignored emp_id 105 in this merge
# See the similarity? It is a inner join.
# Ofcourse you can choose the type of join you want to perform.
```

```python
pd.merge(df4, df5,how='right')
```

```python
data6 = {'emp_id1' : [101,102,103,104],
         'salary' : [234, 345,200,400],
         'pf' : [23,34,56,78],
         }
         
data7  = {'emp_id2' : [101,102,104,105],
          'mob_no' : [3434,5656,767,878],
          'house_nu' : [345,345,345,56]
    
}
```

```python
df6=pd.DataFrame(data6)
df7=pd.DataFrame(data7)
```

```python
df6
```

```python
df7
```

```python
pd.merge(df6,df7 , left_on='emp_id1' , right_on='emp_id2',how = 'inner')
# You could use right_on and left_on to specify the common column explicitly if the names are different
```

```python
df_sales=pd.read_csv('./data fsds/sales_data_final.csv')
df_sales
```

```python
# TODO: add another column to store if the profit is positive or negetive

def profit_flag(a):
    if a>0: return '+'
    elif a<0: return '-'
    else: return '='
df_sales['profit_flag']=df_sales['profit'].apply(profit_flag)
```

```python
df_sales[df_sales['profit_flag']=='='].count()
```

```python
df_sales['len_cust_name']=df_sales['customer_name'].apply(len)
# apply is used to apply function on the row(s) specified. Function can be inbuillt custom or lambda
```
