```python
# Dropping the data as it is only a single data
data_copy=data.copy(deep=True)
print(data_copy[~data_copy.Reviews.str.isnumeric()])
data_copy=data_copy.drop(index= data_copy[~data_copy.Reviews.str.isnumeric()].index[0])
print('_'*23)
print(data_copy[~data_copy.Reviews.str.isnumeric()])

```

```python
# convert the data typing of reviews as it is object now
data_copy['Reviews']=data_copy['Reviews'].astype('int')
```

```python


def to_numeric(raw:str):
    if 'm' in raw.lower():
        return float(raw.replace('m','').replace('M',''))*(1024**2)
    if 'b' in raw.lower():
        return float(raw.replace('b','').replace('G',''))*(1024**3)
    elif 'k' in raw.lower():
        return float(raw.replace('k','').replace('K',''))*(1024)
        
        
data_copy['Size']=data_copy['Size'].apply(to_numeric)
```

```python
import matplotlib.pyplot as plt
plt.hist(data_copy['Size'],bins=100)

# It is a power law distr. so Median is a better fillna
```

```python
data_copy['Size']=data_copy['Size'].fillna(data_copy['Size'].median())
```

```python
data_copy['Size']=data_copy['Size'].astype('int')
```

```python
data_copy.dtypes
```

```python
data_copy['Price($)']=data_copy['Price($)'].str.replace('$','').astype('float')
```

```python
data_copy.dtypes
```

```python
data_copy = data_copy.rename(columns={'Price': 'Price($)'})
```

```python
data_copy
```

```python
# Cleaning Install columns
data_copy['Installs']=data_copy['Installs'].str.replace(',','').str.replace('+','').astype('int')
```

```python
data_copy=data_copy.rename(columns={'Installs':'Installs(+)'})
```
