```python
import pandas as pd 
ap=pd.read_csv('./pandas2_dataset/AgentPerformance (2).csv')
ap.set_index('SL No')
alr=pd.read_csv('./pandas2_dataset/AgentLogingReport.csv')
alr
```

```python
alr['Logout Time']=pd.to_timedelta(alr['Logout Time'])
alr['Login Time']=pd.to_timedelta(alr['Login Time'])
alr['Duration']=pd.to_timedelta(alr['Duration'])
alr['Date']=pd.to_datetime(alr.Date)
```

```python
def fdisplay(inp:pd.DataFrame):
    with pd.option_context('display.max_colwidth', None,
                       'display.max_columns', None,
                       'display.max_rows', None):
        display(inp)
```

```python
fdisplay(alr.sort_values(['Agent','Date']))
```

```python
ap.Date=pd.to_datetime(ap['Date'])
    # ap.Date.dt.dayofweek
ap['weekofyear']=ap['Date'].dt.isocalendar().week
    # ap.groupby('weekofyear')
ap['Average Resolution Time']=pd.to_timedelta(ap['Average Resolution Time'])
ap['Average Response Time']=pd.to_timedelta(ap['Average Response Time'])

ap
```

```python

# weekly Report


import numpy as np
weeklyReport=pd.DataFrame(columns=['Agent','week','lenient_avg','strict_avg','attendance','total_chats','total_feedback','avg_response_time','avg_resolution_time','Total_online_period'])
agents=list(set(ap['Agent Name'].tolist()))
day_count=pd.Series(ap['Date'].unique()).dt.isocalendar().groupby('week').count()['day']
Dates=pd.DataFrame(ap['Date'].unique())


for i in agents:
    
    for wkdyno in day_count.index:
        wkday_count=Dates[Dates[0].dt.isocalendar().day<=5][0].dt.isocalendar().groupby('week').count()['day'].loc[wkdyno]
        weeklyReport.loc[len(weeklyReport.index)]=([
            
            i,  # Agent
           
            wkdyno,  # week
            
            ap[ap['Agent Name']==i][ ap['weekofyear']==wkdyno ][ ap['Total Chats' ]!=0].mean()['Average Rating'], # lenient_avg
            ap[ap['Agent Name']==i][ ap['weekofyear']==wkdyno ].sum()['Average Rating']/max(wkday_count,ap[ap['Agent Name']==i][ ap['weekofyear']==wkdyno ][ ap['Total Chats']!=0].count()['Average Rating']),  #strict_avg
            ap[ap['Agent Name']==i][ ap['weekofyear']==wkdyno ][ ap['Total Chats']!=0].count()['Average Rating'],   #attendance
            ap[ap['Agent Name']==i][ ap['weekofyear']==wkdyno ]['Total Chats'].sum(),   # total_chats
            ap[ap['Agent Name']==i][ ap['weekofyear']==wkdyno ]['Total Feedback'].sum(),
            ap[ap['Agent Name']==i][ ap['weekofyear']==wkdyno ]['Average Response Time'].mean(),
            ap[ap['Agent Name']==i][ ap['weekofyear']==wkdyno ]['Average Resolution Time'].mean(),
            alr[alr['Agent']==i][ap['Date'].dt.weekday==wkdyno]['Duration'].sum()
            ])
weeklyReport
```
