*$\mu$ = Population mean*<br>
*$\sigma$ = Standard Deviation*
# Central Limit Theorem
x $\approx$ N($\mu$ ,  $\sigma$)


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters of the normal distribution
mean = 40  # Mean of the distribution
std_dev = 111  # Standard deviation of the distribution

# Generate random samples from the normal distribution
samples = np.random.normal(mean, std_dev, 1000000)

# Create a histogram of the samples
plt.hist(samples, bins=100, density=True, edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normal Distribution')

# Display the plot
plt.show()
```

If a distribution is plotted whose data points are the mean of all *random samples* you take from a population, It will always be a **gaussian distribution**

```python
Population = samples

sampleset=[np.mean(np.random.choice(Population,1000)) for _ in range(10000)]

plt.hist(sampleset, bins=50, density=True, edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(' Distribution mof mean of all samples')

# Display the plot
plt.show()
```

If your population **doesn't follow gaussian distribution** than the distribution of mean of all samples taken from that non gaussian population **with a sample size of at least 30** will have *log normal distribution*.

```python
Population = np.random.lognormal(0,0.1,104000)

sampleset=[np.mean(np.random.choice(Population,1000)) for _ in range(800)]

plt.hist(sampleset, bins=45,density=True, edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(' Distribution of mean of all samples')

# Display the plot
plt.show()
```

# **Z-Test**

The Z-test is a statistical hypothesis test used to determine whether the means of two populations are significantly different from each other when the population standard deviations are known. It is also applicable for larger sample sizes when the Central Limit Theorem can be applied.

The Z-test follows a normal distribution and compares the means of two groups to assess if they are statistically different. The test statistic used in the Z-test is the Z-score, which is calculated by subtracting the population mean from the sample mean and dividing it by the standard deviation.

**Steps to perform a Z-Test:**

1. **Formulate hypotheses:** State the null hypothesis ($H_0$) and the alternative hypothesis ($H_1$). The null hypothesis assumes that there is no significant difference between the means of the two populations.

2. **Choose significance level:** Select the desired level of significance (e.g., 0.05 or 0.01). It determines the probability of rejecting the null hypothesis when it is true (Type I error).

3. **Collect and analyze data:** Gather data from both populations and calculate the sample means ($\bar{x}_1$ and $\bar{x}_2$) and the population standard deviations ($\sigma_1$ and $\sigma_2$).

4. **Compute the test statistic:** Calculate the Z-score using the formula:
      ### $Z = \frac{{(\bar{x}_1 - \bar{x}_2)}}{{\sqrt{\frac{{\sigma_1^2}}{{n_1}} + \frac{{\sigma_2^2}}{{n_2}}}}}$
where $n_1$ and $n_2$ are the sample sizes of the two populations.

5. **Determine the critical region:** Look up the critical value of the Z-score from the Z-table or use statistical software. The critical region is determined based on the significance level chosen in step 2.

6. **Compare the test statistic with the critical value:** If the absolute value of the calculated Z-score falls in the critical region, reject the null hypothesis. Otherwise, fail to reject the null hypothesis.

7. **Draw conclusions:** Based on the comparison in step 6, conclude whether there is sufficient evidence to support the alternative hypothesis or not.

**Applications of Z-Test:**

The Z-test is commonly used in various fields for hypothesis testing, including:

1. **Biomedical research:** Comparing the effectiveness of different treatments or interventions.

2. **Quality control:** Determining if there is a significant difference in the mean performance of a manufacturing process before and after making improvements.

3. **Market research:** Assessing the impact of marketing strategies on consumer behavior.

4. **Educational research:** Analyzing the effectiveness of different teaching methods.

**Disadvantages of Z-Test:**

1. **Known population standard deviation:** The Z-test requires knowing the standard deviation of the populations being compared, which is often not known in practice. If the standard deviation is unknown, the t-test is more appropriate.

2. **Sample size requirement:** The Z-test assumes a large enough sample size for the Central Limit Theorem to hold. For smaller sample sizes, other tests like the t-test or non-parametric tests may be more suitable.

3. **Limited to means comparison:** The Z-test is specifically designed to compare the means of two populations. It is not applicable for comparing other statistics, such as medians or proportions.

**Example Statement:**

A company is testing the effectiveness of two different advertising strategies to determine which one leads to higher customer engagement. They randomly select two groups of customers: Group A and Group B. They measure the number of clicks on an online advertisement for each group over a one-week period. The collected data is as follows:

Group A: $n_1 = 50, \bar{x}_1 = 120, \sigma_1 = 20$

Group B: $n_2 = 60, \bar{x}_2 = 110, \sigma_2 = 25$

Perform a Z-test at a 0.05 significance level to determine if there is a significant difference in customer engagement between the two advertising strategies.

**Solution:**

1. **Formulate hypotheses:**

   Null hypothesis ($H_0$): There is no significant difference in customer engagement between the two advertising strategies.
   
   Alternative hypothesis ($H_1$): There is a significant difference in customer engagement between the two advertising strategies.

2. **Choose significance level:**

   Significance level: $\alpha = 0.05$

3. **Collect and analyze data:**

   Group A: $n_1 = 50, \bar{x}_1 = 120, \sigma_1 = 20$
   
   Group B: $n_2 = 60, \bar{x}_2 = 110, \sigma_2 = 25$

4. **Compute the test statistic:**

   - Calculate the Z-score using the formula:
   #### $\quad \quad \left[Z = \frac{{(\bar{x}_1 - \bar{x}_2)}}{{\sqrt{\frac{{\sigma_1^2}}{{n_1}} + \frac{{\sigma_2^2}}{{n_2}}}}}\right]$

   - Substituting the given values:
   #### $\quad \quad \left[Z = \frac{{(120 - 110)}}{{\sqrt{\frac{{20^2}}{{50}} + \frac{{25^2}}{{60}}}}}\right]$

   - Calculating the value of Z:
   #### $\quad \quad \left[Z \approx 2.109\right]$

1. **Determine the critical region:**

   Since the significance level is 0.05, we divide it equally between the two tails, resulting in a critical region of 0.025 in each tail. From the Z-table or statistical software, we find the critical Z-value for a two-tailed test at 0.025 significance level to be approximately 1.96.

2. **Compare the test statistic with the critical value:**

   Since the absolute value of the calculated Z-score (2.109) falls in the critical region (greater than 1.96), we reject the null hypothesis.

3. **Draw conclusions:**

   Based on the analysis, we have sufficient evidence to reject the null hypothesis. Therefore, we conclude that there is a significant difference in customer engagement between the two advertising strategies.
