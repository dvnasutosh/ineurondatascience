the normal distribution is characterized by two parameters: the mean (μ) and the standard deviation (σ).

- Normal distribution: continuous probability distribution
- Symmetric bell-shaped curve
- Parameters: mean (μ) and standard deviation (σ)
- Mean represents the central tendency of the distribution
- Standard deviation determines the spread or variability of the data
- Also known as Gaussian distribution
- Commonly used in statistical analysis and modeling
- Many natural phenomena approximate a normal distribution
- Follows the 68-95-99.7 rule (approximately 68% of data falls within one standard deviation, 95% within two standard deviations, and 99.7% within three standard deviations)

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters of the normal distribution
mean = 0  # Mean of the distribution
std_dev = 1  # Standard deviation of the distribution

# Generate random samples from the normal distribution
samples = np.random.normal(mean, std_dev, 100000)

# Create a histogram of the samples
plt.hist(samples, bins=30, density=True, edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normal Distribution')

# Display the plot
plt.show()
```

- Poisson distribution: discrete probability distribution
- Number of events in a fixed interval
- Independent and constant average rate
- Parameter: λ (lambda)- Lambda (λ) in the context of the Poisson distribution represents the average rate of events occurring in a given interval. It determines the shape and characteristics of the distribution. The value of λ specifies the expected number of events that will happen in the interval.
- Probability mass function (PMF)
- Mean and variance equal to λ
- Modeling rare events
- Call center arrivals, manufacturing defects
- Approximation of binomial distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter of the Poisson distribution
lambda_ = 3  # Average rate of events

# Generate random samples from the Poisson distribution
samples = np.random.poisson(lambda_, 1000000000)

# Create a histogram of the samples
plt.hist(samples, bins=20, density=True, edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Poisson Distribution')

# Display the plot
plt.show()
```

The log-normal distribution is a continuous probability distribution of a random variable whose logarithm follows a normal distribution. In other words, if a random variable X follows a log-normal distribution, then the logarithm of X follows a normal distribution.

Here are some key points about the log-normal distribution:

1. Definition:
   - The log-normal distribution describes the distribution of a random variable X, where X > 0.
   - It is characterized by two parameters: μ (mean of the logarithm of X) and σ (standard deviation of the logarithm of X).

2. Probability density function (PDF):
   - The PDF of the log-normal distribution is given by: f(x) = (1 / (xσ√(2π))) * exp(-(ln(x) - μ)^2 / (2σ^2)), where x > 0.

3. Logarithmic transformation:
   - If X follows a log-normal distribution, then Y = ln(X) follows a normal distribution with mean μ and standard deviation σ.

4. Properties:
   - The log-normal distribution is positively skewed (right-skewed) since its logarithm follows a normal distribution.
   - The mean of the log-normal distribution is given by: E(X) = exp(μ + (σ^2) / 2).
   - The variance of the log-normal distribution is given by: Var(X) = [exp(σ^2) - 1] * exp(2μ + σ^2).

5. Applications:
   - The log-normal distribution is commonly used to model data that are skewed and only positive-valued, such as stock prices, income distribution, and sizes of biological organisms.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters of the log-normal distribution
mu = 0  # Mean of the logarithm of X
sigma = 0.41  # Standard deviation of the logarithm of X

# Generate random samples from the log-normal distribution
samples = np.random.lognormal(mu, sigma, 10000)

# Create a histogram of the samples
plt.hist(samples, bins=30, density=True, edgecolor='black')

# Add labels and title
plt.xlabel('Value (X)')
plt.ylabel('Probability Density')
plt.title('Histogram of Log-Normal Distribution')

# Display the plot
plt.show()
```
