```python
# This is a QQ plot 
# If the dataset follow the perfect fit line than it is a normal distribution
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab

def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()

# plot_data(np.log(xlognormal))
plot_data(iris['petal_length'])
plot_data(iris['petal_width'])
plot_data(iris['sepal_width'])
plot_data(iris['sepal_length'])
```
