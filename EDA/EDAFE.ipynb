## **Scoring each entry for possible phishing website**

```python
# Scoring each entry

# Introducing bias or a multiplier set by industry experts to specify the comparative importance of each column

bias= {col: 1 for col in df.columns}


bias
```

```python
# Implementing bias

bias_df= df.copy(deep=True)
for col in bias_df.columns:
    bias_df[col]*=bias[col]

bias_df
```

```python
bias_df['phishing_score']=bias_df.sum(axis=1)
```
