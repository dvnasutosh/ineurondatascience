```python
# q2 : try to print below by using while loop : 
        
# A
# B H 
# C I N
# D J o S
# E K p T W
# F L Q U X z
# G M R V Y 

letterUTF=ord('A')
letter=letterUTF
row=1
col=1
diff=7

while letter<=ord('Z'):
    print(chr(letter), end=' ')
    letter+=(diff-col)
    col+=1
    if col>row:
        letterUTF+=1
        letter=letterUTF
        col=1
        row+=1
        print()
```

```python
import os
import smtplib
import dotenv
dotenv.load_dotenv()
def mailTo(rcv,sub,body,EMAIL_ADDR=os.environ.get('EMAIL'),PASS=os.environ.get('PASSWORD')):
    with smtplib.SMTP("smtp.gmail.com",587) as smtp:
        smtp.ehlo()
        # Identifies with the smtp server we are requesting to connect
        smtp.starttls() # Encrypt our conntection
        smtp.ehlo()     # Re-identify our connection
        
        smtp.login(EMAIL_ADDR,PASS)     # Logging in 

        msg=f'Subject:{sub} \n\n {body}'
        smtp.sendmail(EMAIL_ADDR,rcv,msg) 
```

```python
# q16 : Write a fun which will take input as a dict and give me out as a list of all the values 
# even in case of 2 level nesting it should work . 
def fetchValues(d:dict):
    out=[]
    for i in d.values():
        if i== dict:
            out.append(fetchValues(i))
        else:
            out.append(i)
    return out
```

```python
# q17 : write a function whihc will take multiple list as a input and give me concatnation of all the element as 
# and output
def concatinateAll(*args):
    out=[]
    for i in args:
        out.extend(i)
    return out
```
